# Data Preparation

## Changing Columns Types

```python
df.columns = df.columns.str.replace(r' ', '_', regex=True)
df.columns = df.columns.str.lower()
```

### order_date & ship_date type to Datetime

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
#df['order_date'] = df['order_date'].dt.strftime('%m/%d/%Y')
df['ship_date'] = pd.to_datetime(df['ship_date'], format='%Y-%m-%d')
#df['ship_date'] = df['ship_date'].dt.strftime('%m/%d/%Y')
```

```python
df.info()
```

### sales, Profit, shipping_cost, discount, and aging type to float

sales, profit, shipping_cost

```python
df.rename(columns={
    'sales':'sales($)',
    'profit':'profit($)',
    'shipping_cost':'shipping_cost($)'
}, inplace=True)
```

```python
df[['sales($)', 'profit($)', 'shipping_cost($)']] = df[['sales($)', 'profit($)', 'shipping_cost($)']].replace(r'(\$)', '', regex=True)
```

```python
df.loc[df['sales($)'] == '0.xf', 'sales($)'] = np.nan # There's '0.xf' in sales($)
df.loc[df['shipping_cost($)'] =='test', 'shipping_cost($)'] = np.nan # there's 'test' value in shipping_cost($)
```

```python
df[['sales($)', 'profit($)', 'shipping_cost($)']] = df[['sales($)', 'profit($)', 'shipping_cost($)']].astype(float)
```

discount

```python
df.loc[df['discount'] == 'xxx', 'discount'] = 0.01
```

```python
df['discount'] = df['discount'].astype(float)
```

### Quantity Columns type to int

```python
df.loc[df['quantity'] == 'abc', 'quantity'] = np.nan
```

```python
df.loc[df['quantity'].isnull(), ['sales($)', 'discount', 'profit($)', 'quantity']]
```

quantity have correlation with sales(dollar)/profit(dollar)

```python
df.loc[(df['quantity'].isnull()) & (df['sales($)'] == 72.0), ['quantity']] = 2
df.loc[(df['quantity'].isnull()) & (df['sales($)'] == 54.0), ['quantity']] = 1
```

```python
df['quantity'] = df['quantity'].astype(int)
```
